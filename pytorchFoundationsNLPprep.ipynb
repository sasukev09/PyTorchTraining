```python
tensB, tensB.shape
```

```python
## After transposing tensB, the multiplication works as it now abides by the matrix mul rules
print(f"Original shapes (tensA and tensB): {tensA.shape, tensB.shape}")
print(f"\nNew shapes after transposing tensB: {tensA.shape, tensB.shape}")
print(f"\nAs the new inner dimensions match, we multiply {tensA.shape} @ {tensB.T.shape}")

print(f"\nNow we can multiply and the result is:")
result = torch.matmul(tensA, tensB.T)
print(result)

print(f"\nProduct shape: {result.shape}")
```

```python
## PyTorch tensor & NumPy
import torch
import numpy as np

#Numpys default datatype is float 64
array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array)
array, array.dtype,  tensor
```

```python
#Making a numpy array of type float 32
array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array).type(torch.float32)
array, array.dtype, tensor, tensor.dtype
```

```python
array = array + 2
array, tensor
```

```python
#Tensor to Numpy array
tensor = torch.ones(7)
tensor_to_numpy = tensor.numpy()
tensor, tensor.dtype, tensor_to_numpy, tensor_to_numpy.dtype
```

```python
# checking if they share memory (if they both change when changes are made to either of them)
tensor = tensor + 1
tensor, tensor_to_numpy
# they do not share the same memory
```

```python
# PyTorch Reproducibility(trying to take random out of random)
import torch
tensor_A = torch.rand(2,3)
tensor_B = torch.rand(2,3)

print(f"Tensor A:\n {tensor_A}")
print(f"Tensor B:\n {tensor_B}")
print(f"Does Tensor A equal Tensor B?\n {tensor_A == tensor_B}")
```

```python
# Random but reproducible tensors
import torch

#Set random seed
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)
tensor_C = torch.rand(2,3)

torch.manual_seed(RANDOM_SEED)
tensor_D = torch.rand(2,3)

print(f"Tensor C:\n {tensor_C}")
print(f"Tensor D:\n {tensor_D}")
print(f"Does Tensor C equal Tensor D?\n {tensor_C == tensor_D}")
# random seed only works for one block of code (if using a notebook)
```
